## Part 1: The Gravitational Potential of the Earth
----

The gravitational potential at a distance $r$ from a mass $M$ is given by:

$$\Phi(r) = -\frac{GM}{r}$$

where G is the gravitational constant.

1. Code up a function that takes as an input the mass of a body $M$ , the cartesian location of the
mass $(x_M , y_M)$, and the location of some distant evaluation point $(x, y)$. Have the function
return the potential at the evaluation point. You may consider approximating the Earth as
a point particle, but in which case, be careful! For in that approximation, the the potential
contains a singularity. In our function, use all modern best practices for coding.
E.g. type hints, type checks, pre and post conditions, doc strings, etc.

```python
def potential(M: float, xm: float, ym: float , x: float, y: float) -> float:
    """
    This function will take in the location and mass of a body and the location of some distant point and calculate the gravitational potential on that point from the body

    Input:
    Mass (kg): The mass of a body
    xm, ym (m): The location of the mass in standard coordinates
    x, y (m): The location of a distant point in standard coordinates

    Output:
    The gravitational potential (V): The gravitational potential caused by the body on the distant point
    """
    if not isinstance(M, (float, np.ndarray)):
        raise TypeError("M is not a float")
    elif not isinstance(xm, (float, np.ndarray)):
        raise TypeError("xm is not a float")
    elif not isinstance(ym, (float, np.ndarray)):
        raise TypeError("ym is not a float")
    elif not isinstance(x, (float, np.ndarray)):
        raise TypeError("x is not a float")
    elif not isinstance(y, (float, np.ndarray)):
        raise TypeError("y is not a float")
    elif M <= 0:
        raise ValueError("M must be positive")
    else:
        r = Re + np.sqrt((x-xm)**2 + (y-ym)**2)
        return -G*M/r
```


2. Make a 1D plot of the absolute value of the gravitational potential $|\Phi|$ as a function of distance $x$ from the surface of the Earth, assuming $y = 0$.
Plot out to $1.5\times$ the distance from the Earth to the Moon. Because the potential falls quickly with r, make the y-axis logarithmic to visualize it better.

```python
x = np.linspace(0.0, 1.5*dEm, 1000)
phi = potential(Me, 0.0, 0.0, x, 0.0)
phiabs = np.abs(phi)
fig, ax = plt.subplots()
ax.plot(x, phiabs, linestyle = "--", color = "purple")
plt.yscale("log")
ax.set_xlabel("X (m)")
ax.set_ylabel("The Log of the Potential (V)")
ax.set_title("The Absolute Gravitational Potential as a Function of X distance")
```

## Part 3: The Gravitational Force Field of the Earth-Moon System
----
The gravitational force $\vec{F}$ 
that a mass $M_1$ exerts on a mass $m_2$ is:

$$\vec{F}_{21} = -G \frac{M_1 m_2}{|\vec{r}_{21}|^2} \hat{r}_{21}$$

where $\vec{r}_{21}$ is the displacement vector from $M_1$ to $m_2$.

1. Code up a function that takes as an input the mass of a body $M_1$, 
the mass of a secondary body $m_2$,
the cartesian location of the first mass $(x_1, y_1)$, 
and the cartesian location of the second body $(x_2, y_2)$. 
Have the function return the two gravitational force components, 
$F_x$ and $F_y$, experienced by the mass $m_2$ 
from the force exerted by $M_1$.
Note that the gravitational force contains a singularity, 
and so if you made the approximation that the Earth is a finite point,
you will need to avoid it.
Like for the 1D potential, make sure it follows modern best-coding practices.

```python
def forceofgravity(M1: float, m2: float, x1: float, y1: float, x2: float, y2: float) -> float:
    """
    This function will take in the location and mass of one body and the location and mass of another body and calculate the gravitational force that the first body exerts on the second body

    Input:
    M1 (kg): The mass of the body exerting the force
    m2 (kg): The mass of the body "feeling" the force
    x1, y1 (m): The location of the body exerting the force in standard coordinates
    x2, y2 (m): The location of the body "feeling" the force in standard coordinates

    Output:
    Fx (N): The gravitational potential caused by the first body onto the second body in the x direction
    Fy (N): The gravitational potential caused by the first body onto the second body in the y direction
    """
    if not isinstance(M1, (float, np.ndarray)):
        raise TypeError("M1 is not a float")
    elif not isinstance(m2, (float, np.ndarray)):
        raise TypeError("m2 is not a float")
    elif not isinstance(x1, (float, np.ndarray)):
        raise TypeError("x1 is not a float")
    elif not isinstance(y1, (float, np.ndarray)):
        raise TypeError("y1 is not a float")
    elif not isinstance(x2, (float, np.ndarray)):
        raise TypeError("x2 is not a float")
    elif not isinstance(y2, (float, np.ndarray)):
        raise TypeError("y2 is not a float")
    elif M1 <= 0:
        raise ValueError("M1 must be positive")
    elif m2 <= 0:
        raise ValueError("m2 must be positive")
    else:
        dx = x2 - x1
        dy = y2 - y1
        rabs = abs(Re + np.sqrt((x1-x2)**2 + (y1-y2)**2))
        Fx = -G*M1*m2*dx/rabs**2
        Fy = -G*M1*m2*dy/rabs**2
        return Fx, Fy
    
```

## Part 4: Altitude of the Saturn V Rocket
----
Rockets function by conservation of momentum.
The ejection of fuel backwards propels the rocket forwards.
The change in a rocket's velocity $\Delta v$ as a 
function of time $t$ (under a set of simplifying assumptions) 
is given by the Tsiolkovsky rocket equation:

$$\Delta v(t) = v_e \ln \left(\frac{m_0}{m(t)}\right) - gt$$

where $m_0$ is the initial "wet" mass (fuel + rocket parts + payload), 
$m(t) = m_0 - \dot{m}t$ is the mass at time $t$, 
$\dot{m}$ is the fuel burn rate (which we will assume to be constant), 
$v_e$ is the fuel exhaust velocity, and $g$ is the gravitational acceleration.

You can find the altitude $h$ of the rocket at "burnout" 
(when all the fuel is used up) by integrating the velocity over time:

$$h = \int_{0}^{T} \Delta v(t) \, dt$$

where $T$ is the total burn time of the rocket, 
which by conservation of momentum arguments is:

$$T = \frac{m_0 - m_f}{\dot{m}}$$

where $m_f$ is the final "dry mass" of the rocket, 
once all the fuel is burned.

1. Calculate the burn time $T$ for the first stage of
the new Saturn V rocket that NASA is planning
to use to carry the Apollo capsule.

```python
T = (M0 - Mf)/Mdot
print(T, "s")
```

2. Define a function that returns the change in velocity $\Delta v(t)$.
It should take all necessary arguments, meaning: the time, 
wet mass, dry mass, burn rate, exhaust speed, and gravitational constant $g$. 
Have it return the change in velocity.
Note that the rocket eventually runs out of fuel, 
so you should design your function carefully so that 
$\Delta v$ eventually becomes zero after all fuel is spent.
Again, document and guard your function according to best practices.

```python
def velocity(t: float, M0: float, Mf: float, Mdot: float, Ve: float, g: float) -> float:
    """
    This function calcualtes a rocket's velocity using Tsiolkovsky's equation 

    Inputs:
    t (s): The time over which the velocity is calculated
    M0 (kg): The mass of the rocket with the fuel inside
    Mf (kg): The mass of the rocket without the fuel inside
    Mdot (kg/s): The rate at which fuel is burned
    Ve (m/s): The velocity of fuel exhaust
    g (m/s^2): The gravitational constant

    Output:
    v (m/s): The rockets velocity as a function of time
    """
    if not isinstance(M0, (float, np.floating)):
        raise TypeError("M0 is not a float")
    elif not isinstance(Mf, (float, np.floating)):
        raise TypeError("Mf is not a float")
    elif not isinstance(Mdot, (float, np.floating)):
        raise TypeError("Mdot is not a float")
    elif not isinstance(Ve, (float, np.floating)):
        raise TypeError("Ve is not a float")
    elif not isinstance(g, (float, np.floating)):
        raise TypeError("g is not a float")
    elif t < 0:
        raise ValueError("t must be 0 or greater")
    elif M0 <= 0:
        raise ValueError("M0 must be positive")
    elif Mf <= 0:
        raise ValueError("Mf must be positive")
    else:
        burn_time = (M0 - Mf)/Mdot
        t_used = min(t, burn_time)
        Mt = M0 - Mdot*t_used
        Vt = Ve*np.log(M0/Mt) - g*t_used
        return Vt
```

3. Compute the altitude of the rocket at the end of the burn.
Do this by numerically integrating from $t=0$ (launch) to $t=T$ (burnout).
Use the `scipy.integrate.quad()` function.

```python
Ht, err = quad(velocity, 0, T, args = (M0, Mf, Mdot, Ve, g))
print(Ht, "m")
```
